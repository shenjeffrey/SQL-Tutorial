<strong>Practice Problem:</strong><br>
Write a query that calculates the percentage of all houses completed in the United States represented by each region. Only return results from the year 2000 and later. Hint: There should be four columns of percentages.

<p>SELECT west/(west + south + midwest + northeast)\*100,<br>
          south/(west + south + midwest + northeast)\*100,<br>
          midwest/(west + south + midwest + northeast)\*100,<br>
          northeast/(west + south + midwest + northeast)\*100<br>

<strong>Practice Problem:</strong><br>
Write a query that returns all rows for which Ludacris was a member of the group.

<strong>SQL Answer:</strong>
<p>SELECT \*<br>
FROM tutorial.billboard_top_100_year_end<br>
WHERE "group" ILIKE '%ludacris%'</p>


```python
billboardData[billboardData['group'].str.contains(r'.*Ludacris.*', flags = re.I)].head()
```

<strong>Practice Problem:</strong><br>
Write a query that returns all rows for songs that were on the charts in 2013 and do not contain the letter "a"

<strong>SQL Answer:</strong>
<p>SELECT \*<br>
FROM tutorial.billboard_top_100_year_end<br>
WHERE song_name NOT ILIKE '%a%'</p>

<strong>Practice Problem:</strong><br>
Write a query that lists top 3 songs from each year, starting with the most recent year.

<strong>SQL Answer:</strong>
<p>SELECT \*<br>
FROM tutorial.billboard_top_100_year_end<br>

<strong>Practice Problem:</strong><br>
Write a query that returns songs that ranked between 10 and 20 (inclusive) in 1993, 2003, or 2013. Order the results by year and rank

<strong>SQL Answer:</strong>
<p>SELECT \*<br>
FROM tutorial.billboard_top_100_year_end<br>
WHERE year_rank BETWEEN 10 AND 20 AND year IN (1993, 2003, 2013)<br>
ORDER BY year, year_rank</p>

```python
billboardData[  (billboardData['year_rank'] >= 10) & (billboardData['year_rank'] <= 20) &
                billboardData['year'].isin([1993, 2003, 2013])].head()
```
